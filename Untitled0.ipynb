```python
import backtrader as bt
from pykrx import stock
import pandas as pd
from datetime import datetime
from datetime import timedelta
from backtrader import cerebro
from pandas.io.formats import style
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

class SmaCross(bt.Strategy): # bt.Strategy를 상속한 class로 생성해야 함. (이동평균선)

    params = dict(
        pfast=5,  # period for the fast moving average   5평선
        pslow=20   # period for the slow moving average  20평선
    )

    def log(self, txt, dt=None): # 매수/매도 기록 함수
        ''' Logging function fot this strategy'''
        dt = dt or self.data.datetime[0]
        if isinstance(dt, float):
                dt = bt.num2date(dt)
                print('%s, %s' % (dt.isoformat(), txt))

    def __init__(self):
        sma1 = bt.ind.SMA(period=self.p.pfast)  # fast moving average
        sma2 = bt.ind.SMA(period=self.p.pslow)  # slow moving average
        self.crossover = bt.ind.CrossOver(sma1, sma2)  # crossover signal



    def next(self):  #전략 함수
        global size
        if not self.position: # not in the market # 포지션이 없으면(매수/매도가 없음)
            if self.crossover > 0:  # 단기 이동 평균선이 장기 이동 평균선을 상향 돌파시 매수

                size = int(self.broker.getcash() / self.data.close[0])
                # 최대 구매 가능 개수
                self.buy(price=self.data.close[0], size=size) #매수
                # 매수 size = 구매 개수 설정
                self.log('BUY CREATE, %.2f' % (self.data.close[0]))
                print(size, 'EA')
        else: #포지션이 있는 경우
            if self.crossover < 0:  # 단기 이동 평균선이 장기 이동 평균선을 하향 돌파시 매도
                self.sell(price=self.data.close[0],size=size)
                # 매도
                self.log('SELL CREATE, %.2f' % (self.data.close[0]))
                print(size, 'EA')
#--------------------------------------------------------------------

# cerebro 가져오기
cerebro = bt.Cerebro()

ticker =  "005930"   # 삼성전자 종목코드
# 전체 종목코드와 종목명 가져오기
# df = df.rename(columns={'시가':'open', '고가':'high', '저가':'low', '종가':'close', '거래량':'volume'})
# df["open"]=df["open"].apply(pd.to_numeric,errors="coerce")

from pykrx import stock
df = stock.get_market_ohlcv("2021-12-01", "2022-07-01", ticker)
df = df.drop('등락률', axis=1)
df.columns = ["open","high","low","close","volume"]

data = bt.feeds.PandasData(dataname=df) #데이터 넣어주기
# 데이터 추가
cerebro.adddata(data)

#cerebro 실행
cerebro.addstrategy(SmaCross)  # 투자 전략 추가
cerebro.broker.setcash(10000000)  # 브로거 설정

# 매매 단위 설정하기
cerebro.addsizer(bt.sizers.SizerFix, stake=100)

# 초기 투자금
init_cash = cerebro.broker.getvalue()
cerebro.run()

# 최종 금액
final_cash = cerebro.broker.getvalue()

print("최종 금액 : ", final_cash, "\\")
print("수익률 : ", float(final_cash - init_cash) / float(init_cash) *100., "%")

cerebro.plot()
```
